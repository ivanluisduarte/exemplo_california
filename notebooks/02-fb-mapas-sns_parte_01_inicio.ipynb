```python
# import matplotlib.pyplot as plt
# import pandas as pd
# import seaborn as sns


# funo que plota um scatter plot com hue de numeros continuos|float|reais
def fnc_plot_scatter_hue_continuo(
    # PARAMETROS OBRIGATRIOS
    dataframe: pd.DataFrame, # dataframe com os dados para plotar no grfico
    x: str, # informao do eixo X
    y: str, # informao do eixo Y
    hue: str, # informao que ser diferenciada pelas cores no grfico

    # PARAMETROS OPCIONAIS
    x_label: bool|str= None, # sem label se False, coluna x se None
    y_label: bool|str= None, # sem label se False, coluna y se None
    hue_label: bool|str= None, # sem label se False, coluna hue se None
    title: str= None, # sem ttulo por padro
    size_title: int= 16, # tamanho da fonte do ttulo
    size_eixos: int= 12, # tamanho da fonte dos eixos X e Y
    flg_negrito_title: bool= True, # flag se o ttulo deve ser negrito
    flg_negrito_eixos: bool= True, # flag se o label dos eixos deve ser negrito
    hue_orientation: str= 'vertical', # 'vertical' ou 'horizontal'
    figsize: list|tuple= (8, 6), # tamanho da figura do grfico
    cmap_palette: str= 'coolwarm', # 'coolwarm', 'bwr', 'seismic', 'Reds'
    alpha: float= 1.0, # nvel de opacidade das cores
) -> None:

    '''
        Funo que plota um scatter plot com hue de numeros continuos|float|reais

        Args:
            # PARAMETROS OBRIGATRIOS
            dataframe: pd.DataFrame, # dataframe com os dados para plotar no grfico
            x: str, # informao do eixo X
            y: str, # informao do eixo Y
            hue: str, # informao que ser diferenciada pelas cores no grfico

            # PARAMETROS OPCIONAIS
            x_label: bool|str= None, # sem label se False, coluna x se None
            y_label: bool|str= None, # sem label se False, coluna y se None
            hue_label: bool|str= None, # sem label se False, coluna hue se None
            title: str= None, # sem ttulo por padro
            size_title: int= 16, # tamanho da fonte do ttulo
            size_eixos: int= 12, # tamanho da fonte dos eixos X e Y
            flg_negrito_title: bool= True, # flag se o ttulo deve ser negrito
            flg_negrito_eixos: bool= True, # flag se o label dos eixos deve ser negrito
            hue_orientation: str= 'vertical', # 'vertical' ou 'horizontal'
            figsize: list|tuple= (8, 6), # tamanho da figura do grfico
            cmap_palette: str= 'coolwarm', # 'coolwarm', 'bwr', 'seismic', 'Reds'
            alpha: float= 1.0, # nvel de opacidade das cores

        Returns:
            None: A funo plota um grfico e no retorna nada

        Examples:
            >>> fnc_plot_scatter_hue_continuo(df, 'longitude', 'latitude', 'median_house_value')
    '''

    # <<< Implementao da funo >>>

    # criando figura e eixo
    fig, ax = plt.subplots(figsize=figsize)

    # criando o grfico de disperso
    sns.scatterplot(
        alpha=alpha,
        ax=ax,
        data=dataframe,
        hue=hue,
        palette=cmap_palette,
        x=x,
        y=y,
    )

    ax.get_legend().remove() # remove a legenda automatica


    # normalizando os valores da coluna do hue
    plt_normalize = plt.Normalize(
        vmax=dataframe[hue].max(),
        vmin=dataframe[hue].min(),
    )

    # mapeando as cores de acordo com a normalizacao e a paleta selecionada
    plt_cm_scalar_mappable = plt.cm.ScalarMappable(
        cmap=cmap_palette,
        norm=plt_normalize,
    )



    if hue_label is False:
        hue_label = ''
    else:
        if hue_label is None:
            hue_label = hue

    # cria a barra de cores que substitui a legenda
    cbar = fig.colorbar(
        alpha=alpha,
        ax=ax,
        mappable=plt_cm_scalar_mappable,
        label=hue_label,
        orientation=hue_orientation,
    )

    # Modificando o tamanho e o peso do texto da legenda
    if hue_orientation == 'vertical':
        cbar.ax.yaxis.label.set_size(size_eixos)  # Tamanho da fonte
        cbar.ax.yaxis.label.set_weight('bold' if flg_negrito_eixos else 'normal')  # Negrito
    else: # 'horizontal'
        cbar.ax.xaxis.label.set_size(size_eixos)  # Tamanho da fonte
        cbar.ax.xaxis.label.set_weight('bold' if flg_negrito_eixos else 'normal')  # Negrito

    # preenchendo os titulos
    if x_label is False:
        ax.set_xlabel(None)
    else:
        if x_label is None:
            x_label = x
        ax.set_xlabel(
            xlabel=x_label,
            size=size_eixos,
            weight='bold' if flg_negrito_eixos else 'normal',
        )

    if y_label is False:
        ax.set_ylabel(None)
    else:
        if y_label is None:
            y_label = y
        ax.set_ylabel(
            ylabel=y_label,
            size=size_eixos,
            weight='bold' if flg_negrito_eixos else 'normal',
        )

    if title is not None:
        ax.set_title(
            label=title,
            size=size_title,
            weight='bold' if flg_negrito_title else 'normal',
        )


    plt.show() # apresenta o grfico
```
